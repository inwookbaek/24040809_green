## 2. JDK 설치
- 참고사이트 : https://wikidocs.net/book/15807

### 2.1. OpenJDK 17 설치
- sudo apt update
- sudo apt install openjdk-17-jdk

### 2.2. 설치된 Java 버전확인
- java -version

### 2.3. java설치경로 확인
- 기본경로 : /usr/lib/jvm/java-17-openjdk-amd64/bin/java

### 2.4. JAVA_HOME 환경변수 설정

#### 2.4.1. 환경변수 파일 수정: 
- sudo vi /etc/environment

#### 2.4.2. 파일에 JAVA_HOME 추가:

- 기본값 : PATH="/usr/local/sbin:/usr/local/bin:/usr/sbin:/usr/bin:/sbin:/bin:/usr/games:/usr/local/games:/snap/bin"
- 추가   : JAVA_HOME="/usr/lib/jvm/java-17-openjdk-amd64"

### 2.5. 환경변수 적용 및 확인
- source 명령어를 사용하여 변경된 내용을 적용하고, JAVA_HOME 환경변수가 올바르게 설정되었는지 확인
- source /etc/environment
- echo $JAVA_HOME

### 2.6. Java 삭제 (선택사항)
- 설치된 Java를 제거하려면 다음 명령어를 사용합니다.
- sudo apt remove openjdk-17-jdk

## 3. DOCKER 설치
- https://docs.docker.com/engine/install/ubuntu/

1단계: 기존 패키지 업데이트

sudo apt-get update && sudo apt-get upgrade -y

2단계: 이전 Docker 버전 제거

sudo apt-get remove docker docker-engine docker.io containerd runc -y

3단계: 필수 구성 요소 설치

sudo apt-get install -y \
    ca-certificates \
    curl \
    gnupg \
    lsb-release

4단계: Docker의 공식 GPG 키 추가

sudo mkdir -p /etc/apt/keyrings
curl -fsSL https://download.docker.com/linux/ubuntu/gpg | sudo gpg --dearmor -o /etc/apt/keyrings/docker.gpg

5단계: Docker 저장소 설정

echo \
  "deb [arch=$(dpkg --print-architecture) signed-by=/etc/apt/keyrings/docker.gpg] https://download.docker.com/linux/ubuntu \
  $(lsb_release -cs) stable" | sudo tee /etc/apt/sources.list.d/docker.list > /dev/null

6단계: 패키지 색인 다시 업데이트

sudo apt-get update

7단계: Docker 엔진, CLI 및 Containerd 설치

sudo apt-get install -y docker-ce docker-ce-cli containerd.io docker-buildx-plugin docker-compose-plugin

8단계: Docker 설치 확인

docker --version

9단계: Docker 서비스 시작 및 활성화

sudo systemctl start docker
sudo systemctl enable docker

10단계: sudo 없이 Docker 실행(선택 사항)

sudo usermod -aG docker $USER

변경 사항을 적용하려면 로그아웃했다가 다시 로그인하거나 터미널을 다시 시작.

docker run hello-world

11단계: Docker 설치 테스트

docker run hello-world

추가 참고사항:
Docker 제거

sudo apt-get purge docker-ce docker-ce-cli containerd.io -y
sudo rm -rf /var/lib/docker /var/lib/containerd

Docker 서비스 상태 확인

sudo systemctl status docker

# Docker Command

# 1. Docker란?

- `데이터 또는 프로그램을 격리시키는 기능`을 제공하는 소프트웨어

## 1.1 docker command

### 1.1.1 상위커맨드 / 하위커맨드 /옵션
- docker container [start/stop/create/run/rm/exec/ls/cp/commit] [--name/-e/-p/-v/-d/-i/-t/-a]
- docker image     [pull/rm/ls/build] [-i]
- docker volumn    [create/inspect/ls/prune/rm]  [--name/-a]
- docker network   [connect/disconnect/create/inspect/ls/prune/rm] []

### 1.1.2 옵션

- --name 컨테이너이름
- -p 호스트번호:컨테이너포트번호      : -p = --publish
- -v 호스트디스크:컨테이너디렉토리    : -v = --volumn
- --net=네트워크명                    
- -e 환경변수이름=값                  : -e = --env
- -d : 백그라운드로 실행              : -d = --detach
- -i : 컨테이너에 터미널(키보드) 연결 : -i = --interactive
- -t : 특수키를 사용하도록 함         : -t = --tty
- -help : 도움말
